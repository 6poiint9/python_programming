```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 6))  # Quadratische Grafik

# --- Punkte für Marker (blaue Kästchen) ---
x_marker = [2, 2, 8, 8, 5]
y_marker = [8, 4, 8, 4, 1]

plt.plot(
    x_marker, y_marker,
    's',
    markerfacecolor='blue',
    markeredgecolor='red',
    markeredgewidth=1,
    markersize=14,
    linestyle=''  # Keine Linien, nur Marker
)

# --- Linienverbindungen (grün, gepunktet, dick) ---
linien = [
    ([2, 8], [8, 8]),  # oben
    ([2, 8], [4, 4]),  # unten
    ([2, 2], [4, 8]),  # links
    ([8, 8], [4, 8]),  # rechts
    ([2, 8], [8, 4]),  # Diagonal
    ([2, 8], [4, 8]),  # Diagonal
    ([5, 2], [1, 4]),  # Spitze links
    ([5, 8], [1, 4])   # Spitze rechts
]

for x, y in linien:
    plt.plot(x, y, linestyle=':', color='green', linewidth=2)

# --- Achsen- und Gitter-Einstellungen ---
#plt.xlim(1, 10)
#plt.ylim(1, 10)
#plt.xticks(range(1, 11))
#plt.yticks(range(1, 11))
#plt.grid(True)
#plt.axis('equal')  # Gleiche Skalierung

plt.xticks(range(1, 11))
plt.yticks(range(1, 11))
plt.grid(True)
plt.xlim(0, 11)
plt.ylim(0, 11)

# --- Titel & Achsenbeschriftung ---
plt.title('My nice pyplot graphic')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.show()
```
