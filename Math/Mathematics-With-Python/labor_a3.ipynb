```python
import matplotlib.pyplot as plt

# Main diamond shape
x = [5, 8, 5, 2, 5]
y = [2, 5, 8, 5, 2]

# Create plot
plt.figure(figsize=(6, 6))
plt.plot(x, y, 'k-.', linewidth=2)            # Dashed-dotted black line for diamond
plt.plot(x[:-1], y[:-1], 'ro', markersize=10)  # Red circular markers (excluding last)

# Horizontal dashed-dotted line (y = 5)
plt.plot([2, 8], [5, 5], 'k-.', linewidth=2)   # Match line style
plt.plot(5, 5, 'ro', markersize=10)            # Red dot in the middle

# Customize axes and grid
plt.xticks(range(1, 11))
plt.yticks(range(1, 11))
plt.grid(True)
plt.xlim(0, 11)
plt.ylim(0, 11)

# Labels and title
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("My nice pyplot graphic")

# Show plot
plt.show()
```
