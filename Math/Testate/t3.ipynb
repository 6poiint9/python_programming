```python
import numpy as np
import matplotlib.pyplot as plt

# Original "A"-Linienzüge definieren
lines = [
    [(-2, -3), (0, 6)],   # linke Linie des A
    [(0, 6), (2, -3)],    # rechte Linie des A
    [(-1, 0), (1, 0)]     # Querbalken des A
]

# Rotationsfunktion
def rotate_points(points, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    rotation_matrix = np.array([
        [np.cos(angle_radians), -np.sin(angle_radians)],
        [np.sin(angle_radians),  np.cos(angle_radians)]
    ])
    return [tuple(rotation_matrix @ np.array([x, y])) for (x, y) in points]

# Plot-Vorbereitung
fig, ax = plt.subplots()
ax.set_aspect('equal')  # unverzerrte Darstellung
ax.grid(True)

# Schwarze Linien (Original)
for line in lines:
    xs, ys = zip(*line)
    ax.plot(xs, ys, color='black')

# Rote Linien (Rotiert um 20° gegen den Uhrzeigersinn)
angle = 20
for line in lines:
    rotated_line = rotate_points(line, angle)
    xs, ys = zip(*rotated_line)
    ax.plot(xs, ys, color='red')

# Achsenbereich einstellen für bessere Ansicht
ax.set_xlim(-4, 4)
ax.set_ylim(-5, 7)

plt.title('Skizze des Buchstaben "A" und Rotation um 20°')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Linien für Buchstabe "A"
lines = [
    [(-3, -3), (0, 6)],    # linker Strich
    [(0, 6), (3, -3)],     # rechter Strich
    [(-2, 0), (2, 0)]      # Querbalken
]

# Rotationsfunktion (gegen den Uhrzeigersinn)
def rotate_points(points, angle_degrees):
    angle_radians = np.radians(angle_degrees)
    rotation_matrix = np.array([
        [np.cos(angle_radians), -np.sin(angle_radians)],
        [np.sin(angle_radians),  np.cos(angle_radians)]
    ])
    return [tuple(rotation_matrix @ np.array([x, y])) for (x, y) in points]

# Plot vorbereiten
fig, ax = plt.subplots()
ax.set_aspect('equal')  # unverzerrt
ax.grid(True, which='both')  # Gitter für alle Ticks

# Achsenbereich festlegen
ax.set_xlim(-4, 4)
ax.set_ylim(-5, 7)

# Achsenbeschriftungen für jeden ganzzahligen Wert
ax.set_xticks(np.arange(-4, 5, 1))  # -4 bis 4
ax.set_yticks(np.arange(-5, 8, 1))  # -5 bis 7

# Schwarze Linien
for line in lines:
    xs, ys = zip(*line)
    ax.plot(xs, ys, color='black')

# Rote Linien (20° Rotation)
for line in lines:
    rotated = rotate_points(line, 20)
    xs, ys = zip(*rotated)
    ax.plot(xs, ys, color='red')

#plt.title('Buchstabe "A" mit Rotation (20° gegen Uhrzeigersinn)')
plt.show()
```
