```python
import numpy as np
import matplotlib.pyplot as plt

# Gegebene Punkte (aus dem Bild abgelesen)
x = np.array([1, 2, 3, 4, 5])
y = np.array([0, -0.75, 1, -0.75, 0])

# Erstelle das lineare Gleichungssystem: Vandermonde-Matrix
A = np.vander(x, increasing=False)
# Löse das Gleichungssystem zur Bestimmung der Koeffizienten
coeffs = np.linalg.solve(A, y)

# Erzeuge feine x-Werte für die Plotlinie
x_plot = np.linspace(0, 6, 400)
# Werte des Polynoms mit den bestimmten Koeffizienten
y_plot = np.polyval(coeffs, x_plot)

# Plot
plt.plot(x_plot, y_plot, 'b', label='Polynom')          # Polynomverlauf
plt.plot(x, y, 'r*', markersize=10, label='Stützpunkte') # Rote Sterne
plt.grid(True)                                           # Gitterlinien
plt.axis('equal')                                        # Unverzerrte Darstellung
plt.legend()
plt.title("Polynom durch 5 Punkte")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegebene Punkte
x = np.array([1, 2, 3, 4, 5])
y = np.array([0, 0, 1, 0, 0])

# Erzeuge Vandermonde-Matrix und löse das Gleichungssystem
A = np.vander(x, increasing=False)
coeffs = np.linalg.solve(A, y)

# Erzeuge x-Werte für Plot
x_plot = np.linspace(0, 6, 400)
y_plot = np.polyval(coeffs, x_plot)

# Plot
plt.plot(x_plot, y_plot, 'b', label='Polynom')            # Polynom in Blau
plt.plot(x, y, 'r*', markersize=10, label='Stützpunkte')  # Punkte als rote Sterne
plt.grid(True)                                            # Gitterlinien
plt.axis('equal')                                         # Unverzerrte Darstellung

# Achsenbereich und Ticks
plt.xticks(np.arange(0, 7, 1))                            # x-Achse: 0 bis 6
plt.yticks(np.arange(-1, 3, 1))                           # y-Achse: -1 bis 2
plt.ylim(-1, 2)
plt.xlim(0, 6)

plt.title("Polynom 4. Grades durch fünf Punkte")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Gegebene Punkte
x = np.array([1, 2, 3, 4, 5])
y = np.array([0, 0, 1, 0, 0])

# Lineares Gleichungssystem lösen
A = np.vander(x, increasing=False)
coeffs = np.linalg.solve(A, y)

# Plotdaten erzeugen
x_plot = np.linspace(0, 6, 400)
y_plot = np.polyval(coeffs, x_plot)

# Plot
fig, ax = plt.subplots()
ax.plot(x_plot, y_plot, 'b', label='Polynom')
ax.plot(x, y, 'r*', markersize=10, label='Stützpunkte')

# Gitter und Beschriftung
ax.grid(True)
ax.set_xticks(np.arange(0, 7, 1))
ax.set_yticks(np.arange(-1, 3, 1))
ax.set_xlim(0, 6)
ax.set_ylim(-1, 2)

# Seitenverhältnis 1:1
ax.set_aspect('equal', adjustable='box')

#ax.set_title("Polynom 4. Grades durch fünf Punkte")
ax.legend()
plt.show()
```
