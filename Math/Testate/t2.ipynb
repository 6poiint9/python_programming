```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter t von 0 bis 8π
t = np.linspace(0, 8 * np.pi, 1000)

# Parametrisierung der Kurve
r = np.sqrt(t)
x = r * np.cos(t)
y = r * np.sin(t)

# Plot
fig, ax = plt.subplots()
ax.plot(x, y, color='black')  # Schwarze Kurve wie im Bild

# Achseneinstellungen
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_xticks(np.arange(-5, 6, 1))
ax.set_yticks(np.arange(-5, 6, 1))

# Gitter & Seitenverhältnis
ax.grid(True)
ax.set_aspect('equal', adjustable='box')

# Achsentitel wie in der Skizze (keine expliziten Labels notwendig)
#plt.title("Spiralkurve $c(t) = \\sqrt{t} \\cdot (\\cos(t), \\sin(t))$")

plt.show()
```
