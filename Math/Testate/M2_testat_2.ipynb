```python
import matplotlib.pyplot as plt
import numpy as np

# Punkte f√ºr das Haus vom Nikolaus
points = np.array([
    [0, 0],    # A
    [1, 0],    # B
    [1, 2],    # C
    [0, 2],    # D
    [0.5, 3]   # E (Spitze)
])

# Linienverbindungen (indizieren Punkte)
lines = [
    (0, 1),  # AB
    (1, 2),  # BC
    (2, 3),  # CD
    (3, 0),  # DA
    (3, 1),  # DB
    (0, 2),  # AC
    (3, 4),  # DE
    (2, 4)   # CE
]

# Funktion zum Zeichnen einer Figur
def draw_figure(points, lines, color):
    for start, end in lines:
        plt.plot(
            [points[start][0], points[end][0]],
            [points[start][1], points[end][1]],
            color=color
        )

# Funktion zum Rotieren der Punkte um den Ursprung
def rotate_points(points, angle_deg):
    angle_rad = np.radians(angle_deg)
    rotation_matrix = np.array([
        [np.cos(angle_rad), -np.sin(angle_rad)],
        [np.sin(angle_rad),  np.cos(angle_rad)]
    ])
    return points @ rotation_matrix.T

# Zeichnen
plt.figure(figsize=(6, 6))
draw_figure(points, lines, 'black')                     # Original
rotated = rotate_points(points, 55)                     # Rotation
draw_figure(rotated, lines, 'red')                      # Rotiert

plt.grid(True)
plt.gca().set_aspect('equal')                           # Unverzerrte Darstellung
plt.show()

```
